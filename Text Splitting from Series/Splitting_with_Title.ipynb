```python
# task 1. root_path 내부에 있는 한국 현대소설 작품집 111편 병합본들을 각각의 단편으로 분리하여
# task 2. store_path 경로에 각각의 단편 분권들을 저장하는 코드

# 단, 111편의 소설 목록만을 추출한 spl_file_list.txt 필요. => extract_Number_Title.ipynb 코드에서 추출 가능

root_path = "C:/Users/rkdms/OneDrive_UNIST/바탕 화면/UNIST/2021_01/[DH Lab] 현대소설111/"
store_path = root_path + "split_files/"

# Preprocessing : Check the format of original files
first_piece_num = [1, 15, 26, 36, 46, 53, 62, 73, 84, 89, 90, 98, 110]
piece_num = len(first_piece_num)

last_piece_num = [first_piece_num[i]-1 for i in range(1, piece_num)]
last_piece_num.append(111)


# Making Original file list
ori_file_list = []
for i in range(0, piece_num):
    ori_file_name = "현대소설 111편 " + str(first_piece_num[i]).zfill(3) + ' _ ' + str(last_piece_num[i]).zfill(3)
    ori_file_list.append(ori_file_name)
    
    
# Making Split file list
spl_file_list = open(root_path + 'spl_file_list.txt', 'r')
spl_file_list = [str for str in spl_file_list.readlines()]


# For each Original file
for i in range(0, piece_num):
    index = first_piece_num[i] - 1
    spl_size = last_piece_num[i] - 1
    
    ori_file_path = root_path + str(ori_file_list[i]) + ".txt"
    ori_file = open(ori_file_path, 'r')
    print("\n" + ori_file_list[i])

    while True:
        line = ori_file.readline()
        if not line: break

        spl_file_name = spl_file_list[index].replace("\n", "")

        if spl_file_name in line:
            print(spl_file_name)
            spl_file_path = store_path + spl_file_name + '.txt'
            spl_file = open(spl_file_path, 'w')
            if spl_size != index: 
                index += 1

        spl_file.write(line)

    ori_file.close()
```
