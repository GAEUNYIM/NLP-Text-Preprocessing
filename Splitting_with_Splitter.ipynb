```python
# import sys
# import io
# sys.stdout = io.TextIOWrapper(sys.stdout.detach(), encoding = 'utf-8')
# sys.stderr = io.TextIOWrapper(sys.stderr.detach(), encoding = 'utf-8')


file_original = open("염상섭전집2권 사랑과 죄 OCR 스캔_전처리.txt", 'rt', encoding='utf-8')

splitter = "---------------"

file_chunck_exist = False

while True:
    line = file_original.readline()
    if not line: break
        
    # If line has splitter, create a new file and store the chunck
    if splitter in line:
        print("splitter in line")
        
        # If there was old file_chunck, close it
        if file_chunck_exist == True:
            file_chunck.close()
            
        # Set the tile of file_chunck
        title = file_original.readline().strip('\n')
        print(title + "\n")
        
        file_chunck = open(title + ".txt", 'w', encoding='utf-8')
        file_chunck.write(title + "\n")
        file_chunck_exist = True
        
        line = file_original.readline()
    
    # Write a line into file_chunck
    file_chunck.write(line)
    

file_original.close()
```
